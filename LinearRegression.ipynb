```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

        # Swedish Committee on Analysis of Risk Premium in Motor Insurance

dataset = pd.read_csv("auto_insurance_sweden.csv")

X = dataset["claims"] # number of claims
Y = dataset["payments"] # total payment for all the claims in thousands of Swedish Kronor for geographical zones in Sweden
    # y = mx + b
b = 0    
m = 0    
alpha = 0.0001       # Learning rate
j = 100             # calculates loss every 100th iteration 100 times
loss = np.zeros(j)   # For storing loss values
n = len(X)

for k in range(j):
    for i in range(250):  # Runs gradient descent 100 times 
        yHat = m*X + b                 # predicts y-value
        Dm = (-2/n) * sum(X*(Y-yHat))  # Partial derivitive of the loss function with respect to m
        Db = (-2/n) * sum(Y-yHat)      # Partial derivitive of the loss function with respect to b
        m = m - alpha * Dm   
        b = b - alpha * Db
    
    loss[k] = 1/n * sum(Y - (m*X + b))**2  # Stores loss value


yPred = m*X + b    # Final regression line
plt.scatter(X,Y)
plt.plot(X, yPred, "red")
plt.xlabel("Claim")
plt.ylabel("Payment")

plt.figure()
plt.plot(np.linspace(0,j,j), loss)  # Loss graph
plt.xlabel("Iterations")
plt.ylabel("Loss value")
```
